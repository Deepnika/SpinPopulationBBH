```python
import bz2

files = ["bbh_spin/posterior_samples/posteriors_1.pkl.bz2",
         "bbh_spin/posterior_samples/posteriors_2.pkl.bz2",
         "bbh_spin/posterior_samples/posteriors_zero_1.pkl.bz2",
         "bbh_spin/posterior_samples/posteriors_zero_2.pkl.bz2"]

for filepath in files:
    newfilepath = filepath[:-4]
#     print(newfilepath)
    with open(newfilepath, 'wb') as new_file, open(filepath, 'rb') as file:
        decompressor = bz2.BZ2Decompressor()
        for data in iter(lambda : file.read(100 * 1024), b''):
            new_file.write(decompressor.decompress(data))
```
